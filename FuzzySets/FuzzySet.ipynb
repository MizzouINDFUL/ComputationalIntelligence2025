First, lets create a trapezoidal membership function (versus a zmf, Gaussian, etc.). Note, this fuzzy library does not keep an *analytical representation* of a membership function. Instead, it appears to sample the domain of interest (the x below) and a membership function is specified in terms of its parameters ($[a, b, c, d]$ for a trapezoid) and each sample (x-axis below) has its membership calculated; the result of which is an array of membership values at a specified discrete set of points. This is a finite way to deal with fuzzy sets. We will talk in class about if you want to work with a discrete and finite sampling scheme or if you want to work analytically. 

Note, I leveraged the scikit-fuzzy library and added a few things like shading the sets when we draw them.

```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

# generate trapezoidal membership function
x = np.arange(0, 5.05, 0.1) # the domain is 0 to 5, sampling rate of 0.1
mfx = fuzz.trapmf(x, [2, 2.5, 3, 4.5]) # this function starts at 2, peaks at 2.5, peak ends at 3, and set is done at 4.5

# lets plot it
plt.figure(figsize=(8, 5))
plt.plot(x, mfx, 'k')
plt.fill_between(x,mfx,color="skyblue",alpha=0.4)
plt.ylabel('Fuzzy membership')
plt.xlabel('The domain of interest')
plt.ylim(-0.1, 1.1)
plt.show()
```

Next, lets make a linguistic variable called Age, with the terms $\{Young,Middle,Old\}$.

```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

# generate trapezoidal membership function
x = np.arange(0, 100.5, 1) 
young = fuzz.trapmf(x, [0, 0, 10, 20]) 
middle = fuzz.trapmf(x, [10, 20, 40, 50]) 
old = fuzz.trapmf(x, [30, 50, 100, 100]) 

# lets plot it
plt.figure(figsize=(8, 5))
plt.plot(x,young,'k')
plt.fill_between(x,young,color="skyblue",alpha=0.4)
plt.plot(x,middle,'k')
plt.fill_between(x,middle,color="red",alpha=0.4)
plt.plot(x,old,'k')
plt.fill_between(x,old,color="green",alpha=0.4)
plt.ylabel('Fuzzy membership')
plt.xlabel('The domain of interest')
plt.ylim(-0.1, 1.1)
plt.show()
```

OK, what is the membership grades for a 40+ year old ... (careful there!!!)

```python
a=young[40]
b=middle[40]
c=old[40]

print('The membership grade for a 40 year old is ...')
print( a, b, c )

# lets plot it
plt.figure(figsize=(8, 5))
plt.plot(x,young,'k')
plt.fill_between(x,young,color="skyblue",alpha=0.4)
plt.plot(x,middle,'k')
plt.fill_between(x,middle,color="red",alpha=0.4)
plt.plot(x,old,'k')
plt.fill_between(x,old,color="green",alpha=0.4)
plt.plot([40,40],[0,1],'r')
plt.ylabel('Fuzzy membership')
plt.xlabel('The domain of interest')
plt.ylim(-0.1, 1.1)
plt.show()
```

Again, like in class, you could have calculated that using a trapezoidal membership function
